```python
import csv
filename = 'smalldata.csv'
with open(filename, 'r') as csvfile:
    filereader = csv.reader(csvfile, delimiter=',')
    header = next(filereader)
    for row in filereader:
        print(row[1])
```

```python
duration=[]
filename = 'smalldata.csv'
with open(filename, 'r') as csvfile:
    filereader = csv.reader(csvfile, delimiter=',')
    header = next(filereader)
    for row in filereader:
        duration.append(int(row[0]))
        
print(duration)
```
